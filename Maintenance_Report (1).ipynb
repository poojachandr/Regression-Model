```python
#Encoding code
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['Equ']= le.fit_transform(data["Equipment"])
```

```python
data
```

```python
df2 = data.drop("Equipment",axis =1)
```

```python
df2
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df2.corr(),annot=True)
```

```python
# Splitting the data into training and testing
```

```python
x= df2.drop("Productivity",axis=1)
y = df2["Productivity"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
# Building Linear Regression to predict the productivity of the machine
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print(lm.coef_)
```

```python
print(lm.intercept_)
```

```python
#Predicting the test dataset with linear model
```

```python
y_pred = lm.predict(x_test)
y_pred
```

```python
#checking the model performance
```

```python
from sklearn.metrics import r2_score
print("Accuracy:",r2_score(y_test, y_pred))
```
